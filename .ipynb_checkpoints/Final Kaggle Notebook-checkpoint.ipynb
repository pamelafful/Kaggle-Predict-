```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib.colors import ListedColormap

from sklearn.feature_extraction.text import CountVectorizer

from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn import metrics

from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_moons, make_circles, make_classification

from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
import nltk
from nltk.stem import WordNetLemmatizer 
from sklearn.utils import resample
```

```python
data=pd.read_csv('C://Users//pamela.afful//Desktop//KaggleGithubFolder//train.csv')
test=pd.read_csv('C://Users//pamela.afful//Desktop//KaggleGithubFolder//test.csv')
test.head(20)
```

```python
import re
pd.set_option('display.max_colwidth',-1)

data['message']=data['message'].apply(lambda i: i.lower())



#################### Removing handles and URLs and special characters######################

handles=r'@\w+'
generic_urls=r'http[s]?://(?:[A-Za-z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9A-Fa-f][0-9A-Fa-f]))+'

data['message'] = data['message'].replace(generic_urls, '', regex=True).replace(handles, '', regex=True)
#df['message']=df['message'].replace('rt :', '' )
data['message']=data['message'].replace('rt :', '' )
data['message'].replace('rt ', '' ,inplace=True)
data['message']=data['message'].replace('\\W', ' ',regex=True)

################### Removing Puctuation###############################

import string
print ('Cleaning punctuation...')
def remove_punctuation_numbers(post):
    punc_numbers = string.punctuation + '0123456789'
    return ''.join([l for l in post if l not in punc_numbers])


data['message'] = data['message'].apply(remove_punctuation_numbers)
data.head()


####################Lemmatize words################
lemmatizer=WordNetLemmatizer()
temp=[]
for word in data['message']:
    temp.append(lemmatizer.lemmatize(word))
    
data['lemmatized']=pd.Series(temp)
data
```

```python
################lOOKING AT THE Class IMBALANCE#################
y=data.groupby(data['sentiment']).count()
y=y.reset_index()
y['sentiment_count']=y['message']
y.plot.bar(x='sentiment', y='sentiment_count')
```

```python
X=data['lemmatized']
y=data['sentiment']

vectorizer = CountVectorizer(stop_words="english")
X_vectorized = vectorizer.fit_transform(X)
                             
X_train,X_test,y_train,y_test = train_test_split(X_vectorized,y,test_size=.3,shuffle=True, stratify=y, random_state=11)                             
```

```python
names = ['Nearest Neighbors', 
         'Linear SVM', 'RBF SVM',          
         'Decision Tree', 'Random Forest','AdaBoostClassifier','Logistic Regression']

classifiers = [KNeighborsClassifier(3),
    SVC(kernel="linear", C=0.025),
    SVC(gamma=2, C=1),
    DecisionTreeClassifier(max_depth=5),
    RandomForestClassifier(max_depth=5, n_estimators=10, max_features=1),    
    AdaBoostClassifier(),
  LogisticRegression(multi_class='multinomial')]
```

# Fine Tuning Logistic Regression

```python
param_grid= {'solver':['newton-cg', 'liblinear', 'sag', 'saga'],'C':[1,5,8,10,25]}
grid_log = GridSearchCV( LogisticRegression(multi_class='multinomial'), param_grid, scoring='f1_macro',n_jobs=-1)
grid_log.fit(X_train, y_train)
grid_log.best_params_
```

```python
log=LogisticRegression(C=10,solver='saga')
log.fit(X_train,y_train)
y_pred=log.predict(X_test)
```

```python
#y_pred=#grid_log.predict(X_test)
metrics.f1_score(y_test, y_pred,average='macro')
```
